```python
import pandas as pd
import matplotlib.pyplot as plt
#mport pandas as pd
import glob
import numpy as np
import matplotlib.image as mpimg
from sympy.geometry import *
%matplotlib inline
```

```python
fn='ResultsCOOOLEN.csv'
df=pd.read_csv(fn)
```

```python
N=len(df.X)
x=df.X[0:N-2]
y=df.Y[0:N-2]
```

```python
img = mpimg.imread("CTAXX.jpg")
img2 = mpimg.imread("CTBXX.jpg")
```

```python
fig, axs = plt.subplots(1,figsize=(20,10))
axs.imshow(img)
fig.tight_layout()
```

```python
xn=np.asarray(x)
yn=-np.asarray(y)
ymin=abs(min(yn))
xmax=abs(max(xn))
yn=ymin+yn
xn=xn-xmax
fig, axs = plt.subplots(1,figsize=(20,10))
axs.scatter(xn,yn)
plt.axis('equal')
fig.tight_layout()
```

```python
fig, axs = plt.subplots(1,figsize=(20,10))
axs.imshow(img2)
fig.tight_layout()
```

```python
theta = np.radians(-60)
c, s = np.cos(theta), np.sin(theta)
R = np.array(((c,-s), (s, c)))
print(R)
```

```python
xr=[]
yr=[]
for j in range(len(xn)):
    q=np.array([xn[j], yn[j]])
    v=np.dot(R,q)
    xr.append(v[0])
    yr.append(v[1])
```

```python
fig, axs = plt.subplots(1,figsize=(20,10))
axs.scatter(xr,yr)
xs=xr[0:220]
ys=yr[0:220]
axs.plot(xr[0:220],yr[0:220])
plt.axis('equal')
fig.tight_layout()
```
